```python
import pandas as pd
import math
```

```python
keywords=['Induction stove',
          'Electric car, jeep & van upto 100KW',
          'Electric car, jeep & van > 100KW to <=200KW',
          'Electric car, jeep & van > 200KW to <=300KW',
          'Others (except cycle) motorcycle & scooter with electric motor for propulsion',
          'Electric three wheelers vehicles',
          'Infrared stove',
          
         ]
def contains(description):
    for keyword in keywords:
        if keyword==description:
            return True
    return False         
specified_column = "Description"
```

```python
months=['Shrawan','Bhadra','Asoj','Kartik','Mangsir','Poush','Magh','Falgun','Chaitra','Baishak','Jestha']
dfs=[]

for month in months:
    df=pd.read_excel(f'excel_files/FTS_upto_{month}_7980.xlsx',sheet_name='5_Imports_By_Commodity',header=2)
    df= df[df[specified_column].apply(contains)]
    
    df.insert(loc=0, column='Month', value=month)
    df['Quantity'] = df['Quantity'].apply(lambda x: math.ceil(x)).astype(int)
    
    dfs.append(df)
 
merged_df=pd.concat(dfs)
merged_df.to_excel('filtered_data.xlsx',index=False)
```
