```python
import pandas as pd
```

```python
#dataframe for shrawan
```

```python
df_shrawan=pd.read_excel('customs_excel_files/FTS_upto Shrawan_7879_final.xlsx',sheet_name='5_Imports_By_Commodity',header=2)
```

```python
desiredDescriptions=['AC/DC Flat charger for electric vehicles',
'With only electric motor for propulsion:',
'Electric three wheelers vehicles',
'Electric car, jeep & van upto 100KW',
'Electric car, jeep & van > 150KW to <=200KW',
'Three wheeler with only electric motors for propulsion',
'With electric motor for propulsion',
'With electric motor for propulsion',
'Induction stove',
'Infrared stove',
'Rice cooker',
'Electro-mechanical domestic appliances, ne  , with electric motor',
'Furnaces and ovens functioning by inductio   or dielectric loss']

filtered_df_shrawan = df_shrawan[df_shrawan['Description'].isin(desiredDescriptions)]
filtered_df_shrawan.insert(loc=0, column='Month', value='Shrawan')
```
