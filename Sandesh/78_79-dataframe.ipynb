```python
#Below is the method to create individal dataframe for each month. 
#Here, dataframe for the month shrawan is created.
#Same can be done for other months and all the individual dataframes can be concatenated as one
```

```python
df_shrawan=pd.read_excel('customs_excel_files/Shrawan.xlsx',sheet_name='5_Imports_By_Commodity',header=2)
```

```python
desiredDescriptions=['AC/DC Flat charger for electric vehicles',
'With only electric motor for propulsion:',
'Electric three wheelers vehicles',
'Electric car, jeep & van upto 100KW',
'Electric car, jeep & van > 150KW to <=200KW',
'Three wheeler with only electric motors for propulsion',
'With electric motor for propulsion',
'With electric motor for propulsion',
'Induction stove',
'Infrared stove',
'Rice cooker',
'Electro-mechanical domestic appliances, ne  , with electric motor',
'Furnaces and ovens functioning by inductio   or dielectric loss']

filtered_df_shrawan = df_shrawan[df_shrawan['Description'].isin(desiredDescriptions)]
filtered_df_shrawan.insert(loc=0, column='Month', value='Shrawan')
```

```python
#Below is the method where all the dataframes are created at once and finally concatenated
```

```python
months=['Shrawan','Bhadra','Asoj','Kartik','Mangsir','Poush','Magh','Falgun','Chaitra','Baishak','Jestha']
dfs=[]

for month in months:
    df=pd.read_excel(f'customs_excel_files/{month}.xlsx',sheet_name='5_Imports_By_Commodity',header=2)
    df = df[df['Description'].isin(desiredDescriptions)]
    df.insert(loc=0, column='Month', value=month)
    dfs.append(df)
merged_df=pd.concat(dfs)
```

```python
merged_df.to_csv('test1.csv',index=False)
```
